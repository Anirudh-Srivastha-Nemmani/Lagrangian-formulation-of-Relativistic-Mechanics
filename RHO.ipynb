```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(u,t):
  return (u[1],-1*u[0]*((1 - (u[1])**2)**1.5))

#with initial condition x(t=0)=0
x0 = [0.5,0]
ts = np.linspace(0,20,400)
us = odeint(f,x0,ts)
xs = us[:,0]

#with initial condition x(t=0)=1
x1 = [1,0]
ts = np.linspace(0,20,400)
us1 = odeint(f,x1,ts)
xs1 = us1[:,0]

#with initial condition x(t=0)=2
x2 = [2,0]
ts = np.linspace(0,20,400)
us2 = odeint(f,x2,ts)
xs2 = us2[:,0]

#with initial condition x(t=0)=4
x3 = [4,0]
ts = np.linspace(0,20,400)
us3 = odeint(f,x3,ts)
xs3 = us3[:,0]

#plotting graphs
plt.rc('axes', labelsize=15)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.axhline(y=0, color='k', linestyle='dotted')
plt.rcParams['axes.xmargin'] = 0
plt.plot(ts,xs,'b', label ='x₀ = 0.5')
plt.plot(ts,xs1,'g', label ='x₀ = 1')
plt.plot(ts,xs2,'y', label ='x₀ = 2')
plt.plot(ts,xs3,'r', label ='x₀ = 4')

#legend plotting
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',frameon=False , borderaxespad=0.)

#printing the plot
plt.show
plt.savefig('fig2.png', bbox_inches="tight")
```
